```python
SYSTEMS = ['ETHYL','ISOPROPYL','ISOBUTYL','TERTBUTYL','NEOPENTYL','CYCLOHEXYL','CYCLOHEXENYL',
           'CYCLOHEXADIENYL']

MOLECULES = ['Mol_01','Mol_02','Mol_03','Mol_04','Mol_05']
for isystem in SYSTEMS:
    DIR_OUT = '../'+isystem+'/'+isystem+'_DATA/'+isystem+'_ZPE.dat'
    OUT_FILE = open(DIR_OUT,"w")
    print(OUT_FILE)
    for imol in MOLECULES:
        DIR = '../'+isystem+'/'+isystem+'_OPT_FILES/'
        FILENAME = DIR+imol+'.log'
        file = open(FILENAME,"r")
        for line in file:
            if re.search('Zero-point correction=',line):
                OUT_FILE.write((line.split()[2]))
                OUT_FILE.write("\n")
                print(line.split()[2])
    OUT_FILE.close()
                
```
