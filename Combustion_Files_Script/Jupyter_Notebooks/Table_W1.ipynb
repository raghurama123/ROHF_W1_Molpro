```python
import re
import os
import numpy as np
import pandas as pd
from pandas import DataFrame as df

au2kcal = 627.50957099203276
```

```python
DIR_CSV = '../CSV_FILES/'
```

```python
SYSTEMS =['ETHYL','ISOPROPYL','ISOBUTYL','TERTBUTYL','NEOPENTYL','CYCLOHEXYL','CYCLOHEXENYL',
           'CYCLOHEXADIENYL']
GS = ['R+O2','ROO','QOOH','cy-Ether+OH','Alkene+OOH']
TS = ['TS_01','TS_02','TS_03','TS_04']
SPECIES =['R+O2','ROO','QOOH','cy-Ether+OH','Alkene+OOH','TS_01','TS_02','TS_03','TS_04']
VALUES = np.zeros((len(SPECIES),len(SYSTEMS)))
TABLE_2 = pd.DataFrame(VALUES,SPECIES,SYSTEMS)
for isystem in SYSTEMS:
    DATA = pd.read_csv(DIR_CSV+isystem+'_W1_Rxn_E.csv',index_col='SPECIES')
    DATA_TS =pd.read_csv(DIR_CSV+isystem+'_TS_W1_Rxn_E.csv',index_col='SPECIES')
    for index in TABLE_2.index:
        if index in GS:
            TABLE_2[isystem].loc[index] = round(DATA['W1'].loc[index],2)
        else:
            TABLE_2[isystem].loc[index] = round(DATA_TS['W1'].loc[index],2)

TABLE_2.index.name='SPECIES'
TABLE_2.to_csv('TABLE_W1.csv')
#TABLE_2.to_csv('TABLE_2_overleaf_style.csv', sep ='&')
```
